HTMLから取得する

```python
from lxml import html
import regex
p0 = regex.compile(r'[\p{Script=Hiragana}\p{Script=Katakana}]+')
```

```python
def parse_html(fname):
    with open(fname,'r',encoding='utf-8') as f:
        lines="".join([l.rstrip() for l in f])

    root=html.fromstring(lines)


    lst=[]
    #words = root.xpath('//div[@id="content"]/div/ul/li[@class="onyomi"]/a')
    words = root.xpath('//div[@id="content"]/div/ul/li/a')
    for word in words:
        if len(word.text)==1 and not p0.match(word.text):
            lst.append(word.text)

    return list(set(lst))
```

```python
syo=parse_html('school1')
```

```python
len(syo)
```

```python
with open('syogaku.dat','w',encoding='utf-8') as f:
    for s in syo:
        f.write(f'{s}\n')
```

```python
cyu=parse_html('school2')
```
