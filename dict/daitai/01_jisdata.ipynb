整形する

```python
from lxml import html
```

```python
def parse_html(fname):
    with open(fname,'r',encoding='utf-8') as f:
        lines="".join([l.rstrip() for l in f])

    root=html.fromstring(lines)


    lst=[]
    words = root.xpath('//div[@class="parts_box"]/ul[@class="search_parts"]/li/a')
    for word in words:
        lst.append(word.text)

    return lst
```

```python
dai1=parse_html('data/jisdai1')
```

```python
dai2=parse_html('data/jisdai2')
```

```python
dai3=parse_html('data/jisdai3')
```

```python
dai4=parse_html('data/jisdai4')
```

保存する

```python
def save_file(fname,lst):
    with open(fname,'w',encoding='utf-8') as f:
        for l in lst:
            f.write(f'{l}\n')
    
```

```python
save_file('jisdai1.dat',dai1)
```

```python
save_file('jisdai2.dat',dai2)
```

```python
save_file('jisdai3.dat',dai3)
```

```python
save_file('jisdai4.dat',dai4)
```
