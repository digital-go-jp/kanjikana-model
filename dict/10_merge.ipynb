```python
import copy
import json
```

```python
def merge(dic1,dic2):
    dic=copy.deepcopy(dic1)
    for kanji,vv in dic2.items():
        if kanji not in dic:
            dic[kanji]=vv
            continue
        for kana,attr in vv.items():
            if kana not in dic[kanji]:
                dic[kanji][kana]=attr
                continue

            dic[kanji][kana]['dics']+=attr['dics']
    return dic
```

```python
def load(fname):
    jdata={}
    try:
        with open(fname,'r',encoding='utf-8') as f:
            jdata = json.load(f)
        return jdata
    except:
        pass
    return jdata
```

```python
jlis=load('jlis2.json')
```

```python
oss=load('oss2.json')
```

```python
crawl=load('crawl2.json')
```

```python
reliable=merge(jlis,oss)
```

```python
reliable=merge(reliable,crawl)
```

```python
with open('reliable.json','w',encoding='utf-8') as f:
    json.dump(reliable,f,indent=2,ensure_ascii=False)
```
