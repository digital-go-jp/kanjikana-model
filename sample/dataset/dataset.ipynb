```python
import pandas as pd

df=pd.read_csv('eng-fra.txt', header=None, names=['eng','fra'], sep='\t')

df.head()
```

```python
sdf = df.sample(frac=1, random_state=0)
sdf.head()
```

```python
siz=len(sdf)//10


val=sdf[:siz]
test=sdf[siz:siz*2]
train=sdf[siz*2:]

train.head()
```

```python
import json

def to_jsonl(df,fname):
    jdata=[]
    for i in range(len(df)):
        eng=df.iloc[i]['eng']
        fra=df.iloc[i]['fra']
        jdata.append({'translation':{'eng':eng, 'fra':fra}})

    with open(fname,'w',encoding='utf-8') as f:
        for l in jdata:
            f.write(json.dumps(l,ensure_ascii=False)+'\n')
```

```python
to_jsonl(train,'train.jsonl')
to_jsonl(val,'val.jsonl')
to_jsonl(test,'test.jsonl')
```
