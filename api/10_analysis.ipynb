```python
seiri=[]
kanji=[]
kana=[]
flg=[]

with open('data/input.txt','r',encoding='utf-8') as f:
    for l in f:
        items = l.rstrip().split(',')
        seiri.append(items[0])
        kanji.append(items[1])
        kana.append(items[2])
        flg.append(items[3])


    
```

```python
dic={}
for kj, kn ,fl in zip(kanji,kana,flg):
    key = f'{kj},{kn}'
    if key not in dic:
        dic[key]=[]
    dic[key].append(int(fl))
```

```python
len(dic)
```

```python
import glob
import datetime

kanji,kana,stat=[],[],[]
stdate,eddate=[],[]
flag=[]
for fname in glob.glob('data/out*'):
    with open(fname,'r',encoding='utf-8') as f:
        for l in f:
            items = l.rstrip().split(',')
            kanji.append(items[0])
            kana.append(items[1])
            stat.append(int(items[2]))
            st=datetime.datetime.strptime(items[-2],'%Y-%m-%d %H:%M:%S.%f')
            ed=datetime.datetime.strptime(items[-1],'%Y-%m-%d %H:%M:%S.%f')
            stepoch=st.timestamp()*1000
            edepoch=ed.timestamp()*1000
            stdate.append(stepoch)
            eddate.append(edepoch)
            key = f'{items[0]},{items[1]}'
            fg = -1
            if key in dic:
                if len(dic[key])==1:
                    fg = dic[key][0]
                
            flag.append(fg)
            
```

```python
import pandas as pd

df = pd.DataFrame({'kanji':kanji,"kana":kana,'stat':stat,'stepoch':stdate,'edepoch':eddate,"flag":flag})

okdf=df[(df['stat']>=50) & (df['flag']!=-1)]
ngdf=df[(df['stat']<50) & (df['flag']!=-1)]
```

```python
print(f'ok={len(okdf)}')
print(f'ng={len(ngdf)}')
```

```python
import numpy as np

okst=okdf['stepoch'].to_numpy()
oked=okdf['edepoch'].to_numpy()

oktim=oked-okst
```

```python
np.average(oktim)
```

```python
np.max(oktim)
```
